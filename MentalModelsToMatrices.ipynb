```python
@backoff.on_exception(backoff.expo, (openai.RateLimitError, openai.APIConnectionError, openai.Timeout), max_tries=20)
def get_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    embeddings = client.embeddings.create(input = [text], model=model)
    return embeddings.data[0].embedding
```

```python

##### set threshold #####
threshold = 0.85
#####

def extract_variables(relationship):
    var1, var2_with_symbol = relationship.split("-->")
    var1 = var1.strip()
    var1 = var1.lower()
    symbol = ""
    if "(+)" in var2_with_symbol:
        _ , var2 = var2_with_symbol.split("(+)")
        symbol = "(+)"

    elif "(-)" in var2_with_symbol:
        _, var2 = var2_with_symbol.split("(-)")
        symbol = "(-)"
    var2 = var2.strip()
    var2 = var2.lower()
    var1 = re.sub(r'[!.,;:]', '', var1)
    var2 = re.sub(r'[!.,;:]', '', var2)
    return var1, var2, symbol

def compute_similarities(variables):
    embedding_list = np.array([get_embedding(x) for x in variables])
    normalized_embeddings = embedding_list / np.linalg.norm(embedding_list, axis=1, keepdims=True)
    similarity_matrix = np.dot(normalized_embeddings, normalized_embeddings.T)
    similar_variables = []
    h,w = similarity_matrix.shape
    assert h == w, "Similarity matrix is not symmetric!"
    for i in range(h):
        for j in range(i+1,h):
            group = set()
            score = similarity_matrix[i][j]
            if score >= threshold:
                group.add(variables[i])
                group.add(variables[j])
            if group:
                sorted_group = tuple(sorted(group))
                similar_variables.append(sorted_group)

    similar_variables = list(set(similar_variables))
    if len(similar_variables) == 0:
        return None
    else:
        return similar_variables

def get_variables(text_list):
    ppls = {}
    variables = set()
    for text in text_list:
        person, relationship = text.strip().split("\n",1)
        ppls[person] = relationship
        rels = relationship.split("\n")
        result_list = [r.strip().split(".",1)[1].strip() for r in rels]
        for i,line in enumerate(result_list):
            variable1, variable2, _ = extract_variables(line)
            variables.add(variable1)
            variables.add(variable2)
    return variables, ppls

def matrify(variable_to_idx, text):
    relationships = text.split("\n")
    relationships = [rels.strip().split(".",1)[1].strip() for rels in relationships]
    variable_list = [extract_variables(relationship) for relationship in relationships]
    matrix = np.zeros((len(variable_to_idx), len(variable_to_idx)), dtype= int)

    for val in variable_list:
        var1, var2 , sym = val
        matrix[variable_to_idx[var1]][variable_to_idx[var2]] = 1 if sym == "(+)" else -1

    return matrix
```

```python


import sys

text_list = txt.split("\n\n")
vars, ppls = get_variables(text_list)
similar_vars = compute_similarities(list(vars))
var_to_new_var = {}
# variable_to_idx = dict(zip(vars, list(range(len(vars)))))

for var in similar_vars:
    print(Fore.BLUE+f"These two variables are similar: {var}")
    while True:
      choice = input(Fore.BLUE+"Merge these two variables into one? (Y/N): ")
      if choice.lower() == "n":
          break
      elif choice.lower() == "y":
          # new_name = input("Enter new name for these variables: ")
          if len(var[0]) >= len(var[1]):
              new_name = var[1]
          else:
              new_name = var[0]
          var_to_new_var[var[0]] = new_name
          var_to_new_var[var[1]] = new_name
          break
      else:
          print(Fore.BLUE+"Neither Y nor N. Please retry...")

# print(f"Old relationships: {txt}")
new_txt = txt
for old_name, new_name in var_to_new_var.items():
    new_txt = new_txt.replace(old_name, new_name)
new_result_list = new_txt.split('\n\n')
print(f"New relationships: {new_result_list}")
new_vars, new_ppls = get_variables(new_result_list)
variable_to_idx = dict(zip(sorted(new_vars), list(range(len(new_vars)))))
for ppl, rels in new_ppls.items():
    matrix = pd.DataFrame(matrify(variable_to_idx, rels), columns= list(sorted(new_vars)), index = list(sorted(new_vars)))
    matrix.to_csv(f"{ppl[:-1]}.csv")
    # print("For ",ppl)
    # print(matrix)

print(Fore.BLUE+Style.BRIGHT+f"Variable to index: {variable_to_idx}")
```
